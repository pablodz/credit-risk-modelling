```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def bernoulli_trial(number, proba):
    n_success = 0
    for i in range(number):
        random_number = np.random.random()
        if random_number < proba:
            n_success += 1
        else:
            pass
    return n_success

def get_ecdf(array_data):
    n = len(array_data)
    x = np.sort(array_data)
    y = np.arange(1, n+1) / n
    return x, y
```

```python
n_defaults = np.empty(1000)

for i in range(1000):
    n_defaults[i] = bernoulli_trial(number=100,
                                    proba=0.05)

bins_centered = np.arange(0, max(n_defaults) + 1.5) - 0.5

_ = plt.hist(n_defaults, density=True, bins=bins_centered,ec='black',alpha=0.8,color='red')
_ = plt.title("Bernoulli trials")
_ = plt.xlabel('number of defaults out of 100 loans')
_ = plt.ylabel('probability')

plt.show()
```

```python

x, y = get_ecdf(n_defaults)

plt.plot(x, y, marker='.', linestyle='none',color='red')
plt.title("probability of losing money distribution")
plt.xlabel('x')
plt.ylabel('y')
plt.show()

# Sum of n_defaults iterations with more than 10
n_lose_money = np.sum(n_defaults >= 10)
# Propbaility of lose money with that restriction
print('Probability of losing money =', n_lose_money / len(n_defaults))
```
